```python
# Ulepszona wizualizacja layout turbin z kolorami i etykietami
import numpy as np
import matplotlib.pyplot as plt

# Dane turbin
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])
labels = ['T1', 'T2', 'T3', 'T4']
colors = ['blue', 'red', 'green', 'yellow']

# Tworzymy wykres
plt.figure(figsize=(8, 6))

# Rysujemy kaÅ¼dÄ… turbinÄ™ z innym kolorem
for i in range(len(x)):
    plt.scatter(x[i], y[i], s=200, c=colors[i], label=labels[i], edgecolors='black', linewidth=2)
    # Dodajemy etykiety dla kaÅ¼dej turbiny
    plt.text(x[i], y[i] + 50, labels[i], ha='center', fontsize=12, fontweight='bold')

plt.title("Layout turbin â€“ Zadanie 2.1 z kolorami i etykietami", fontsize=14, fontweight='bold')
plt.xlabel("x [m]", fontsize=12)
plt.ylabel("y [m]", fontsize=12)
plt.grid(True, alpha=0.3)
plt.axis("equal")
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
